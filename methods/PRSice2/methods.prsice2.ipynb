```python
import os
import pandas as pd
import numpy as np
import warnings
warnings.filterwarnings("ignore")
```

```python
def PRSice2(gwas_path:str,
            disease:str,
            clump_r2=0.1,
            ckb='',
            output_prefix=''):

    os.system(f"Rscript PRSice.R \
                    --prsice PRSice_linux \
                    --base {gwas_path} \
                    --target ../../target_data/target.data.impute \
                    --binary-target T \
                    --clump-r2 {clump_r2} \
                    --pheno ../../results/{disease}/samples.txt \
                    --keep ../../results/{disease}/samples.txt \
                    --cov ../../target_data/covar_list.tsv \
                    --cov-col @PC[1-10],Sex,YoB \
                    --beta \
                    --print-snp \
                    --score sum \
                    --seed 68 \
                    --out ../../results/{disease}/PRSice2{ckb}.{disease}.clumpr2.{clump_r2}")
```

```python
def PRSice2_new(gwas_path:str,
            disease:str,
            clump_r2=0.1,
            output_prefix='',
            covar_cols='@PC[1-10],Sex,Age'):

    os.system(f"Rscript PRSice.R \
                    --prsice PRSice_linux \
                    --base {gwas_path} \
                    --target ../../target_data/target.data.impute\
                    --binary-target T \
                    --clump-r2 {clump_r2} \
                    --pheno ../../results/{disease}/samples.txt \
                    --keep ../../results/{disease}/samples.txt \
                    --cov ../../target_data/covar_list.tsv \
                    --cov-col {covar_cols} \
                    --beta \
                    --print-snp \
                    --score sum \
                    --seed 68 \
                    --out ../../results/{disease}/{output_prefix}")
```

```python
clump_r2_list = [0.1, 0.4, 0.7, 0.9]
```

## 1. Breast cancer

```python
# for gwas EAS (BBJ+)
covar_name = ['age', 'no.age']
for c in clump_r2_list:
        for i, covar_cols in enumerate(['@PC[1-10],Sex,Age', '@PC[1-10],Sex']):
                prefix = f'PRSice2.BC.clumpr2.{c}.{covar_name[i]}'
                PRSice2_new(gwas_path='../../sum_stats_data/breast_cancer/hum0197.v3.BBJ.BC.v1/gwas.PRSice.txt',
                                disease='breast_cancer', clump_r2=c, output_prefix=prefix, covar_cols=covar_cols)
```
